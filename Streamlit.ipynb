```python
%%writefile app.py
import streamlit as st
import numpy as np
import pandas as pd
from PIL import Image
import time

#display the simple text
st.title("Hello World")

#create a Data Frame
df = pd.DataFrame({
    'first_column':[1,2,3,4],
    'second_column':[10,20,30,40]
})

#display the Data Frame
st.write("Here is my Data Frame")
st.write(df)

#create a line chart
chart_data = pd.DataFrame(
    np.random.randn(20,3),columns=['a','b','c']
)
st.line_chart(chart_data)
```

```python
%%writefile widgets.py
import streamlit as st
import numpy as np
import pandas as pd
from PIL import Image
import time


## write code here
st.title("Streamlit Text Input")

name=st.text_input("Enter your name:")
if name:
  st.write(f"Hello,{name}")

age=st.slider("Enter your age:-",0,100,25)
if age:
  st.write(f"Your age is {age}")

option=["Python","C++","Java","C"]
choice=st.selectbox("Select your choice",option)
if choice:
  st.write(f"Your choice is {choice}")

df = pd.DataFrame({
    'first_column':[1,2,3,4],
    'second_column':[10,20,30,40]
})
df.to_csv("Sample.csv")
st.write("Here is my Data Frame")
st.write(df)

uploaded_file = st.file_uploader("Choose a file",type="csv")
if uploaded_file:
  df=pd.read_csv(uploaded_file)
  st.write(df)
```
