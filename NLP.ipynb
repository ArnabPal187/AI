**Tokenization Example****

```python
corpus="""Hello, Welcome to NLP . I am vk18'fan . I am CR7 fan!"""
```

```python
print(corpus)
```

```python
## sentences->paragraphs
from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentences in documents:
  print(sentences,end='\n')
```

```python
##paragraph->words
from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
```

```python
words
```

```python
##sentence->words
for sentences in documents:
  print(word_tokenize(sentences))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus) #treated punctuation as seperate word ex-'
```

```python
#full stop(.) is not treated as seperate word it is add with previous word
# but only last fullstop treated as seperate word
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
