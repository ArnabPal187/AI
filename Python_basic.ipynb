```python
##Variable lenght arguments
## Positional arguments
def print_numbers(*args):
  for numbers in args:
    print(numbers,end=" ")
print_numbers(1,2,3,4,5,"Arnab")
print('\n')
## Keywords arguments
def print_details(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")
print_details(name="Arnab",age=32,grade=True)
```

```python
def print_details1(*args,**kwargs):
  for val in args:
    print (f"Positional argument:{val}")
  for key,value in kwargs.items():
    print(f"{key}:{value}")
print_details1(1,2,3,4,"Arnab",name="Arnab",age=32,grade=True)
```
