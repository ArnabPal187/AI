```python
## break exit current loop
j=0
for i in range(5):
  while j<=5:
    j=j+1
    if(j==2):
      print("Break statement execute and out of j loop")
      break
    print("The valur of i={},j={} in while loop".format(i,j))
  print("The valur of i={},j={} in for loop".format(i,j))
```

```python
## The continue statement skips the current iteration and continues with he next
for i in range(10):
  if i%2==0:
    continue
  print(i,end=" ")
```

```python
## pass is a null operation and do nothing
for i in range(5):
  if i==3:
    pass
  print(i,end=" ")
```

```python
for i in range(3):
  for j in range(3):
    print(f"i:{i} and j:{j}")
```
